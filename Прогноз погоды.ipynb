```python
import pandas as pd #импорт библиотеки pandas
import numpy as np # импорт библиотеки numpy
import datetime as dt# импорт библиотеки datetime

utc_time = dt.datetime.utcnow()# всемирное координированное время 
period = dt.timedelta(hours = 7)# поправка к всемирному координированному времени.
local_time = utc_time + period# местные дата и время
print(local_time)

df = pd.read_csv('C:/Users/nkalinin/Documents/temp.csv',delimiter =';',encoding='1251',header = 0)# читаем с помощью pd.read_csv
df = df.dropna()# удаляем строку с значениями NaN
df['2016'] =df['2016'].astype('float')# преобразуем колонку в числа с плавающей запятой
df['2020'] =df['2020'].astype('float')# преобразуем колонку в числа с плавающей запятой
print(df)# выводим на печать преобразованный df
a = str(local_time)# локальные  дату и времяпреобразуем в строку
a = int(a[8]+a[9])# выбираем число месяца
a = a - 1# т.к строки начинаются с индексом 0 вычитаем 1
print(a)
df1 = df.iloc[a]# строка данных температуры за n-е число месяца за 9 лет
data_mean = round(np.mean(df1),1)# среднее значение температуры за n-е число месяца  за 9 лет
data_std = round(df1.std(),1)# среднеквадратичное отклонение температуры
print(data_mean)
print(data_std)
print(data_mean/data_std)
print(df1)
```
